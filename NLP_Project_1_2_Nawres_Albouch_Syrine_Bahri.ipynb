# Cleaning

```python
def get_special_characters(data):
  special_characters = []
  for char in data:
    if char.isalnum() == False and char not in special_characters:
      special_characters.append(char)
  return special_characters
```

```python
special_characters  = []
p_tashkeel = re.compile(r'[\u0617-\u061A\u064B-\u0652-\u202e]')
for i in df.index:
 text = df['Text'][i]
 text = re.sub(p_tashkeel,"", text)
 special_characters += get_special_characters(text)
```

```python
set(special_characters)
```

```python
#using regular experssions to clean 
data = []
special_characters = []
for i in df.index:
 text = df['Text'][i]
 text = re.sub(p_tashkeel,"", text)
 text = re.sub(' +',' ',text)
 text = re.sub(r"['_×‘؛%٪#✽\*'ٰ]",'',text)
 text = re.sub('"','',text)
 text = re.sub('@','',text)
 text = re.sub(r'\+','',text)
 text = re.sub(r'[0-9]','',text)
 text = re.sub(r'[\{\u202b}\{\u200b}\{\u200f}\{\u200e}\{\u200c}\{\x7f}\{\u202c}\{\xad}]','',text) 
 data.append(text)
```

## Creating a Dataset of sentences

```python
dataset = []
for text in data:
  dataset += re.split(r'[,;؟،«.»/:)(\-!–]',text)
```

```python
sentences = []
for text in dataset[100000:]:
  text = " ".join([w for w in text.split(" ") if not w in arb_stopwords])
  text = text.strip()
  if(len(text.split(' '))>3):
    sentences.append(text)
```

```python
#arabic language lemmatizer
def lemmatization(sentence) :
  lemmer = qalsadi.lemmatizer.Lemmatizer()
  words = sentence.split(' ')
  for i in range(len(words)):
    lemma = lemmer.lemmatize(words[i])
    words[i] = lemma
  return ' '.join(words)
```

```python
def get_lemmatized_dataset(data):
  for i in range(len(data)):
    data[i] = lemmatization(data[i])
    if(i % 100 == 0):
      print(i)
  return data
```

```python
#lemmetize each word in each sentence
sentences = get_lemmatized_dataset(sentences)
```
