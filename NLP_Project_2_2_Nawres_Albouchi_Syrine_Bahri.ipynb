# Language Modeling

```python
def get_co_occurence(data,word1,word2):
  co_occurence = 0
  for sentence in data :
    if word1 in sentence and word2 in sentence:
      co_occurence += 1
  return co_occurence
```

```python
def get_semantic_validity_coefficient(data,sentence,word):
  try:
    freq = []
    for w in sentence:
      freq.append(get_co_occurence(data,w,word))
    mean = (1/len(freq))*sum(freq)
    #print(mean)
    std = ((1/len(freq))*sum([(x - mean)**2 for x in freq]))**0.5
    return mean/std
  except:
    print('division par zero')
```

```python
def lemmatization(sentence) :
  lemmer = qalsadi.lemmatizer.Lemmatizer()
  words = sentence.split(' ')
  for i in range(len(words)):
    lemma = lemmer.lemmatize(words[i])
    words[i] = lemma
  return ' '.join(words)
```

```python
def language_modeling(data_,sentence_):
  try : 
    index = -1
    sentence = lemmatization(sentence_).split()
    n = len(sentence)
    average = 0
    j = -1
    for word in sentence:
      #print(word)
      average += get_semantic_validity_coefficient(data_,sentence,word)
    average /= n
    #print(average)
    for word in sentence:  
      j += 1
      semantic_validity_coefficient = get_semantic_validity_coefficient(data_,sentence,word)
      #print(word,semantic_validity_coefficient)
      if semantic_validity_coefficient < average:
        index = j
  except :
    return -1
  return index   
  
```

```python
def get_accuracy_of_language_modeling(test,data_,test_size=48):
  accuracy = 0
  for i in test_df.index:
    sentence = test_df['text'][i]
    label = test_df['Label'][i]
    predict = language_modeling(data_,sentence)
    if predict == label:
      accuracy += 1
  return accuracy/test_size

```

## Cosine similarity Formula

```python
# A et B deux vecteurs de memes dimentions !!!!!!
def cosine_sim(A,B):
# compute cosine similarity
  cosine = np.dot(A,B)/(norm(A)*norm(B))
  return cosine
```

##1. AraBERT

```python
# Load the Arabert model
model = AutoModel.from_pretrained("aubmindlab/bert-base-arabert")
# Access the embedding layer
embedding_layer = model.embeddings

# Load the BERT base model tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
def get_arabert_embeddings(sentence):
  embeddings=[]
  sentence = sentence.split()
  for i in sentence:
    input_ids = tokenizer.encode(i, return_tensors='pt')
    embeddings.append(embedding_layer(input_ids))
  return embeddings
```

```python
def pca(embeddings,k,n):
  pca = PCA(n_components=n)
  a = pca.fit(embeddings[k][0].detach().numpy())
  return a.explained_variance_ratio_
```

```python
def get_list_pca(embeddings):
  data = []
  n = []
  result = -1
  for i in range(len(embeddings)):
    n.append(embeddings[i].shape[1])
  n = min(n)
  for i in range(len(embeddings)):
    data.append(pca(embeddings,i,n))
  return data,n
```

```python
def arabert(sentence):
  embeddings = get_arabert_embeddings(sentence)
  data , n = get_list_pca(embeddings)
  moy = np.zeros(n)
  result = -1
  for i in range(len(data)):    
    moy = np.add(moy,data[i])  
  moy /= len(embeddings)
  for i in range(len(embeddings)) :
    if i == 0 :
      context = data[i+1]
      if cosine_sim(data[i],np.divide(moy,context)) < 0.45 : 
        result = i
    else :
      context = data[i-1]
      if cosine_sim(data[i],np.divide(moy,context)) < 0.45 :    
        result = i
  return result 
```

```python
def get_arabert_accuracy(test,test_size=48):
  accuracy = 0
  test_df['Arabert']=0
  for i in test_df.index:
    sentence = test_df['text'][i]
    label = test_df['Label'][i]
    predict =  arabert(sentence)
    if predict != -1:
      test_df['Arabert'][i] = sentence.split()[predict]
    else : 
      test_df['Arabert'][i] =""
    if predict == label:
      accuracy += 1
  return accuracy/test_size
```

```python
def get_fasttext_embeddings(sentence):
  embeddings=[]
  sentence = sentence.split()
  for i in sentence:
    embeddings.append(fasttext_model[i])
  return embeddings
```

```python
def fasttext(sentence,n=300):
  embeddings = get_fasttext_embeddings(sentence)
  average = np.zeros(n)
  result = -1
  for i in range(len(embeddings)):    
    average = np.add(average,embeddings[i])  
  average /= len(embeddings)
  for i in range(len(embeddings)) :
    if i == 0 :
      context = embeddings[i+1]
      if cosine_sim(embeddings[i],np.divide(average,context)) < 0.45 : 
        result = i
    else :
      context = embeddings[i-1]
      if cosine_sim(embeddings[i],np.divide(average,context)) < 0.45 :    
        result = i
  return result
```

## Concatenating FastText-Word2vec-Arabert-TF-HUB Multi

```python
def most_frequent(List):
	counter = 0
	num = List[0]
	
	for i in List:
		curr_frequency = List.count(i)
		if(curr_frequency> counter):
			counter = curr_frequency
			num = i

	return num
```

```python
def models_concatenating(sentence):
  predict_arabert = arabert(sentence)
  predict_fasttext = fasttext(sentence)
  prdict_word2vec = word2vec(sentence)
  predict_tf_TF_HUB = TF_HUB(sentence)
  predictions = [predict_arabert,predict_fasttext,prdict_word2vec ,predict_tf_TF_HUB]
  return  most_frequent(predictions)
```

```python
def get_models_accuracy_(test,test_size=38):
  accuracy = 0
  test_df['final_result'] = 0
  for i in test_df.index:
    sentence = test_df['text'][i]
    label = test_df['Label'][i]
    predict = models_concatenating(sentence)
    if predict != -1:
      test_df['final_result'][i] = sentence.split()[predict]
    else : 
      test_df['final_result'][i] =""
    if predict == label:
      accuracy += 1
  return accuracy/test_size
```
